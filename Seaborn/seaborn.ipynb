Seaborn
- is a statistical plotting library
- uses matplotlib
- well designed to work with pandas dataframes

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # built in data-set
tips.head()
```

```python
# dist plot - allows us to show the distribution of univariate (one variable) set of observations
sns.distplot(tips['total_bill']) # takes one column only
```

```python
# the line over histogram is kde and we can remove that by
sns.distplot(tips['total_bill'], kde=False)
```

```python
# you can change number of bins by
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
# joint plot - allows you to match two dist plots with two different variables
sns.jointplot(x= "total_bill", y= "tip", data=tips, kind='hex')
# x=column 1 you want to plot, y=column 2 you want to plot, data=dataset, kind= what type of joint plot you want (default - scatter, other options - hex, reg, kde)
```

```python
# pair plot - plots pair wise relationships across an entire data frame for numerical columns
# - joint plot for every single possible combination of the numerical columns in given data frame
# - nice way to visualize data at once

sns.pairplot(tips) # takes dataset
```

```python
# can take hue - which is a categorical column name and then pair plot differentiates data in plot according to that categorical values
sns.pairplot(tips, hue='sex')
```

```python
# rug plot - draws a dash mark for every point on uniform or unique variant distribution (one variable)
sns.rugplot(tips['total_bill'])

# difference between histogram and rig plot is that histogram has bins and it counts how many dashes (that are on rug plot) were in that bin and then it shows it as number on y axis
```

```python
sns.distplot(tips['total_bill'])
```

```python
# kde line is built based off of rug plot

# kde plot - kernal density estimation plot, sum the normal distribution to get kde
```

```python
# Plotting categorical data

# bar plot - plot that allows you to aggreagte the categorical data based off of some fucntion and by default thats the mean
sns.barplot(x='sex', y='total_bill', data=tips)
# x= some categorical column, y = some numerical column, data= dataset
```

```python
# instead of mean if you want to see division based on std deviation then 
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# count plot - same as bar plot except estimator is expicitly counting the number of occurrences
# only x value needed and dataset
sns.countplot(x='sex', data=tips)
```

```python
# box plots - these types of plots are used to show that distribution of categorical data 
# box plot is also known as box and whisker plot and it shows distribution of quantative data in a way that hopefully facilitates comparision between the variables.

sns.boxplot(x='day', y='total_bill', data=tips)
# x=categorical, y=numerical, data=dataset
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
# now we will see box plot division by day and smoker too

# tells things like - people on friday tend to have a larger bill if they are not a smoker
```

```python
# violen plots - plays a similar role to box plot

sns.violinplot(x='day', y='total_bill', data=tips)
# x=categorical, y=numerical, data=dataset

# violen plot unlike box plot allows us to actually plit all the components that correspond to the actual data points
# and its essentially showing the kernal density of the underlying distribution
# and it is showing the distribution of points on its side and this allows you to reveal more information about the data points themselves
# difficult to read compared to box plots, but gives more information
```

```python
# hue to add another column
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# we can have one half show one hue and another half show another
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)

# gives more information about the comparision of data points
```

```python
# strip plot - draws a scatter plot where one variable is categorical

sns.stripplot(x='day', y='total_bill', data=tips)

# gives scatter plot based off category
```

```python
# many points are stacked on top of each other so you cant really see, for this
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)

# jitter adds random noise to seperate some of these stacked points, you can see the density a little better by this
```

```python
# you can add hue in strip plot too
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# you can also call split on this
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# combining the idea of split plot and violen plot to create a swarm plot

# swarm plot - similar to strip plot except that the points are adjusted so that they dont overlap

sns.swarmplot(x='day', y='total_bill', data=tips)

# one drawback - even though theyre pretty nifty and cool and show lot of information; sometimes they 
# dont actually scale that well to very arge numbers both in terms of ability to show all the points 
# (these points just go way too wide since you dont want to stack anything on top of each other) 
# and it also takes a lot of computation to arrange them 
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# you can plot swarm plot on top of a violen plot (this is a better option than using swarm plot seperately)
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips)

# can be used for exploratary data analysis
```

```python
# Factor plot - most general form of all these plots
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')

# most general because you just need to give the kind and ten it will create that type of plot for you
```

```python
# Matrix Plots (heat maps)

import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
# Heat maps 
# - in order for it to work properly your data should already be in matrix form

# rows and columns should have same index type or should have some correlation
# the tips data right now has indexes on rows but columns dont have any such things
# we can use corr function to create a matrix like

# corr only takes numeric data so need to do this
numeric_data = tips.select_dtypes(include=[np.number])
numeric_data.corr()
```

```python
# creating a heat map
tips_corr = numeric_data.corr()
sns.heatmap(tips_corr)

# heat map really just colors in values according to gradient scale
# for larger data matric you can tell just by looking at colors what the realtion between values is
```

```python
# you can annotate the cell values with the color they belong to
sns.heatmap(tips_corr, annot=True)
```

```python
# you can specify color map too
sns.heatmap(tips_corr, annot=True, cmap='coolwarm')
```

```python
flights
# this data isnt in any matrix form either
```

```python
# we can use pivot table to convert any data to matrix form 
# we specify what value we want as index and what value as column and what value as matrix value
fp = flights.pivot_table(index='month', columns='year', values='passengers')
# here we will have months on rows, year on columns and passengers as the matrix values

# now we have values that correlate with both rows and columns 
```

```python
sns.heatmap(fp, cmap='magma')
# here we can see just by colors which year and month had highest and lowest flights
# flights increase with increasing year and most flights are around jun, july, aug months
# 1960 - jul,aug had the most flights according to heat map
```

```python
sns.heatmap(fp, cmap='coolwarm', linewidths=2, linecolor='white')
```

```python
# Cluster Map
# - uses hierarchal clustering to produce a clustered version of heat map
# - clusters rows and columns together based off their similarity

sns.clustermap(fp)
# you can see months and year are not in order now because they are placed in a way so they are next to or close to the months/year they have most similarity with
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
# if you want to normalzie the scale then 
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
# standard_scale=1 will normalize the scale from 0-1

# you can see nov,jan,feb are low passenger months
# and july,aug are high passenger months
```

```python
# Grids - to automate subplots based off of features in our data

import seaborn as sns
%matplotlib inline

iris =sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# we can customize the pairplot using pair grid with a lot more control over the actual plot

sns.PairGrid(iris)
# gives us an empty pair grid, takes all numerical columns and gives us grid
```

```python
g = sns.PairGrid(iris)
# on this grid map the plot you want

import matplotlib.pyplot as plt
g.map(plt.scatter) # this will plot scatter plot on the grids
```

```python
# to plot specific plots on specific parts of grid- pass that plot types function
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# Facet Grid

tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
# pass the column and row name here
# we get an empty grid
# now we can plot on these grids
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill') # dist plot on total bill column
```

```python
# for scatter
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

```python
# Regression Plots

# LM plot - allows you to display linear models with seaborn

import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
# gives a scatter plot with a linear fit on top of it
```

```python
# seperation using hue
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# change markers
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
# if you want to change the sixe of markers
# seaborn is using matplotlib under the hood so if you want to go and change scatter plot of matplotlib then use core function (this function depends on case to case bases so you need documentation)
# in this case we want to change scatter plot marker size so we use scatter_kws and pass in a dictionary to increase size to 100
# this is actually just a call to matplotlib to cahnge underlying stuff of plot that seaborn is using 
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

```python
# use grid instead of seperating with hue
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
# now you have two seperate grids each representing one type of sex
```

```python
# if you want to add anoter varibale, add it as time
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
# if you want another variable info, add hue 
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex')
# size and aspect of this arent good
```

```python
# to adjust aspect 
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6)
```

```python
# Style and Color

import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x='sex', data=tips)
```

```python
# set style for all plots
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
# remove spines
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```

```python
# size and aspect
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)

# seaborn is calling matplotlib so when its figsize is changed seaborn plots size is also changed
```

```python
# scale and context

# for example you need the plot for some poster
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```

```python
# if you still want to increase font size then
sns.set_context('poster', font_scale=3) # 3 times of default
sns.countplot(x='sex', data=tips)
```
