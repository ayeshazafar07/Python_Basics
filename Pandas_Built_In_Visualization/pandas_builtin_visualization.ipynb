Pandas own visualization capabilities that are built based on matplotlib

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df1['A'].hist() 
# its calling matplotlib under the hood so you can pass matplotlib arguments
```

```python
df1['A'].hist(bins=30) 
```

```python
# plots dont look nice so import seaborn and then plots will look like seaborn plots
import seaborn as sns
df1['A'].hist(bins=30) 
```

```python
df1['A'].plot(kind='hist', bins=30)
# you can change kind here 
```

```python
df1['A'].plot.hist()
```

```python
df1.plot.line(x=df1.index,y='B', figsize=(12,3), lw=2)
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
# we have information of three columns now
# A and B on x and y axis and then C is represented by color
```

```python
# instead of color you can use size to represent another column in plot
df1.plot.scatter(x='A', y='B', s=df1['C']*70) # *70 because otherwise size will be small
```
