Matplotlib
- Most popular plotting library of python 
- Gives control over every aspect of a figure or plot
- similar to MatLab
- Works well with pandas and numpy

```python
import matplotlib.pyplot as plt
```

```python
# to see the plots inside the jupyter notebook 
%matplotlib inline

# or we just use plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

Two ways of creating a matplotlib plot
1. Fucntional method
2. Object Oriented method

```python
# Functional Method
plt.plot(x,y)
plt.show() 
```

```python
# labels
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
# multi plots 
plt.subplot(1,2,1) # subplot (rows=1, columns=2, plot number you are refering to=1)
plt.plot(x,y,'r') # r = color red

plt.subplot(1,2,2) # subplot (rows=1, columns=2, plot number you are refering to=2)
plt.plot(y,x,'b') # b= color blue 
```

```python
# Object Oriented Method
# main idea is to create fig object and then just call methods on it

fig = plt.figure() # figure object is created

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
# add a set of axes to this figure by passing a list 
# add_axes (left, bottom, width, heigth)
```

```python
# plotting those axes

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes.plot(x,y)

axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title("Title 1")
axes2.set_title("Title 2")
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3) 
plt.tight_layout() # fixing overlap of plots issue
```

```python
# sub plots

fig,axes = plt.subplots(nrows=1, ncols=2) # you can add number of rows and columns

# axes is just a list of matplotlib axes

# instead of plot we can iterate axes like
#for current_xs in axes:
#    current_xs.plot(x,y)

# we can also index it
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("Title")
axes[1].set_title("Title")

plt.tight_layout()
 
```

```python
# Figure Size, Aspect Ratio and DPI
fig = plt.figure(figsize=(3,2), dpi=100) 
# figsize (width, height)
# dpi = dots per inch or pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# saving a figure

fig.savefig("my_picture.png", dpi=100)
```

```python
# two plots on same figure
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
# adding a legend
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# for legend you need to give label
ax.plot(x,x**2, label='x square')
ax.plot(x,x**3, label='x cube')

ax.legend()
# ax.legend(loc=0) 
# giving location where you want legend
# check code from documentation
```

```python
# setting colors
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange') # basic color strings or rgb color hex codes
```

```python
# line width and transparency
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=3, alpha=0.5) 
# 1 is default width
# alpha = line transparency
```

```python
# line style
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw=3, linestyle='--')
# linewidth = lw 
# linestyle = ls
# linestyle options = '--', '-.', ':', 'steps'
```

```python
# markers

# marking where the x points occur on the plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw=1, ls='-',marker='o', markersize=15,
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='red')
# marker options - 'o', '+', '*', '1', '2'
```

```python
# only showing plot between certain values of x or y
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw=3, linestyle='--')


ax.set_xlim([0,1]) # pass the list of x values between which you want to see the plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', lw=3, linestyle='--')
ax.set_ylim([0,1])
```

```python
# scatter plot
plt.scatter(x,y)
```

```python
# histogram
from random import sample
data = sample(range(1,1000), 100)
plt.hist(data)
```

```python
# box plot
data = [np.random.normal(0 , std, 100) for std in range(1,4)]

plt.boxplot(data, vert=True, patch_artist=True)
```
